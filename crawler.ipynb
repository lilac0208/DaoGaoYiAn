```python
import  json, ssl, urllib.request

url = 'https://od.moi.gov.tw/MOI/v1/pbs'
context = ssl._create_unverified_context()

with urllib.request.urlopen(url, context=context) as jsondata:
    #將JSON進行UTF-8的BOM解碼，並把解碼後的資料載入JSON陣列中
     data = json.loads(jsondata.read().decode('utf-8')) 

for i in data:
      print(i['srcdetail'],'\t',i['areaNm'],'\t',i['direction'],'\t',i['y1'],'\t',i['happentime'],'\t',i['roadtype'],'\t',i['road'],'\t',i['modDttm'],'\t',i['comment'],'\t',i['happendate'],'\t',i['x1'],'\t')

Saving = []

for i in range(50):


      lng_val = None
      lat_val = None

      try:
        if 'x1' in data[i] and data[i]['x1'] != '':
            lng_val = float(data[i]['x1'])
        else:
            print(f"Warning: data[{i}]['x1'] is missing or empty, setting lng to None.")
            
        if 'y1' in data[i] and data[i]['y1'] != '':
            lat_val = float(data[i]['y1'])
        else:
            print(f"Warning: data[{i}]['y1'] is missing or empty, setting lat to None.")
            
      except ValueError as e:
        print(f"Error converting coordinates for data[{i}]: {e}. Setting lng/lat to None.")
        # 如果轉換失敗，則保持為 None
        lng_val = None
        lat_val = None
      except KeyError as e:
        print(f"Error: Missing key {e} in data[{i}]. Setting lng/lat to None.")
        lng_val = None
        lat_val = None
      inccident = {    
           "srcdetail" :data[i]['srcdetail'],
            "areaNm" :data[i]['areaNm'],
           "direction" :data[i]['direction'],
            "lng" :lng_val,
            "lat" :lat_val,
           "roadtype" :data[i]['roadtype'],
           "modDttm" :data[i]['modDttm'],
            "comment" :data[i]['comment']
            
            
            }
      Saving.append(inccident)   
      
with open("data.json" , "w",encoding="utf-8" ) as f:
      json.dump(Saving , f  ,ensure_ascii=False, indent=1) 



```
